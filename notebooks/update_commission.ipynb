```python
import json
from pathlib import Path

import pandas as pd
import requests

from config import Config
```

```python
base_dir = Path(".").resolve()
data_dir = base_dir / "data"
```

```python
csv_file = data_dir / "to_base_credentials_test.csv"

df = pd.read_csv(csv_file)

for index, row in df.iterrows():
    credential_id = row["credential_id"]
    markup = row["markup_after_action"]

    url = f"http://distributor.rules.api.pro.logitravel.internal/api/organizations/lgt/agencies/{credential_id}/products/hot/comissions"
    headers = {
        "Content-Type": "application/json",
        "Accept": "text/plain",
        "Authorization": f"ApiKey {Config.APIKEY}",
    }
    data = {"value": markup}

    try:
        response = requests.put(url, headers=headers, json=data)

        if response.status_code == 200:
            print(response.json())
        elif response.status_code == 404:
            print("Error: Agencies not found.")
            print(None)
        else:
            print(f"Error: Received unexpected status code {response.status_code}.")
            print(None)

    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        print(None)
```
